# Fronteira de Possibilidades de Produção

Em economia, a **fronteira de possibilidades de produção (FPP)**, também designada por curva de possibilidade de produção (CPP), ilustra graficamente a escassez dos fatores de produção e cria um limite para a capacidade produtiva de uma empresa, país ou sociedade.

Assim na função supõe-se a produção de apenas dois bens, sabendo que os restantes fatores de produção já são dados.

A Função FPP tem a forma de curva e, ao longo desta, é possível estabelecer diferentes pontos máximos de eficiência produtiva e respectivos trade-offs resultantes do aumento/diminuição da produção de um dos bens expostos.

Qualquer ponto no seu interior representa ineficiência produtiva, pois não se produz na máxima eficiência possível, estabelecida pela fronteira. Qualquer ponto no seu exterior é impossível como prova a lógica. 

__[wikipedia](https://pt.wikipedia.org/wiki/Fronteira_de_possibilidades_de_produ%C3%A7%C3%A3o)__

## Modelo 1 FPP 

$$ Comida = \text{constante} + (\alpha \times \text{Roupa}) - (\text{Roupa}^{\beta}) $$

```python
def f_comida(constante, alpha, beta, roupa):
    return constante + (alpha * roupa) - (roupa ** beta)
```

```python
# Definir os valores de roupa
roupas = range(0,13)

# Definir as constantes a ser testadas
con1 = 40
con2 = 50
con3 = 30

# Definir os valores de comida para cada conjunto de parámetros
comidas_1 = [f_comida(constante=con1, alpha=0.2, beta=1.6, roupa=r) for r in roupas]
comidas_2 = [f_comida(constante=con2, alpha=0.2, beta=1.6, roupa=r) for r in roupas]
comidas_3 = [f_comida(constante=con3, alpha=0.2, beta=1.6, roupa=r) for r in roupas]

# Definir os valores de comida para roupa=8
roupa_alvo = 8
com1 = f_comida(constante=con1, alpha=0.2, beta=1.6, roupa=roupa_alvo)
com2 = f_comida(constante=con2, alpha=0.2, beta=1.6, roupa=roupa_alvo)
com3 = f_comida(constante=con3, alpha=0.2, beta=1.6, roupa=roupa_alvo)

# Represente um ponto fora da curva comidas_3 dentro de comidas_2
roupa_alvo_2 = 3
comida_alvo_2 = f_comida(constante=con2, alpha=0.2, beta=1.6, roupa=roupa_alvo_2) - 1

# Definir o texto com os resultados
r1 = f'c={con1}, Comida(roupa={roupa_alvo})={com1:.2f}'
r2 = f'c={con2}, Comida(roupa={roupa_alvo})={com2:.2f}'
r3 = f'c={con3}, Comida(roupa={roupa_alvo})={com3:.2f}'

r4 = f'roupa={roupa_alvo_2}, comida={comida_alvo_2:.2f}'

# Inicializar nova figura
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

# Adicionar primeira curva 
ax.plot(roupas, comidas_1, color="red", label=r1);

# Adicionar resultado 
ax.scatter(roupa_alvo, com1, color="red" );

# Adicionar todos os resultados
ax.plot(roupas, comidas_2, color="green", label=r2);
ax.scatter(roupa_alvo, com2, color="green" );

ax.plot(roupas, comidas_3, color="blue", label=r3);
ax.scatter(roupa_alvo, com3, color="blue" );

# Adicionar ponto fora da curva mais próxima da origem
ax.scatter(roupa_alvo_2, comida_alvo_2, marker="s", color="black", label=r4);

# Configurar presentação 
ax.grid();
ax.set_xlabel("Roupa");
ax.set_ylabel("Comida");
ax.set_ylim([0, max(comidas_1 + comidas_2 + comidas_3)]);
ax.set_xlim([0, max(roupas)]);
ax.set_title('Evolução da FPP');
ax.legend();
```

## Modelo 2 FPP 

$$ Comida = 100 + (\alpha \times \text{Roupa}) - (\text{Roupa}^{\beta}) $$

```python
def f_comida_2(alpha, beta, roupa):
    return f_comida(constante=100, alpha=alpha, beta=beta, roupa=roupa)
```

```python
# Definir os valores de roupa
roupas = np.arange(0,11, 0.5)

beta1 = 2
beta2 = 2.5
beta3 = 2.2

# Definir os valores de comida para cada conjunto de parámetros
comidas_1 = [f_comida_2(alpha=0.2, beta=beta1, roupa=r) for r in roupas]
comidas_2 = [f_comida_2(alpha=0.2, beta=beta2, roupa=r) for r in roupas]
comidas_3 = [f_comida_2(alpha=0.2, beta=beta3, roupa=r) for r in roupas]

# Definir os valores de comida para roupa=8
roupa_alvo = 6
com1 = f_comida_2(alpha=0.2, beta=beta1, roupa=roupa_alvo)
com2 = f_comida_2(alpha=0.2, beta=beta2, roupa=roupa_alvo)
com3 = f_comida_2(alpha=0.2, beta=beta3, roupa=roupa_alvo)

# Definir o texto com os resultados
r1 = f'beta={beta1}, Comida(roupa={roupa_alvo})={com1:.2f}'
r2 = f'beta={beta2}, Comida(roupa={roupa_alvo})={com2:.2f}'
r3 = f'beta={beta3}, Comida(roupa={roupa_alvo})={com3:.2f}'

# Analisis do custo de oportunidade
roupa_alvo_2 = 5.5
comida_alvo_2a = f_comida_2(alpha=0.2, beta=beta1, roupa=roupa_alvo_2)
comida_alvo_2b = f_comida_2(alpha=0.2, beta=beta2, roupa=roupa_alvo_2)

r4 = f'roupa={roupa_alvo_2}, comida={comida_alvo_2a:.2f}'
r5 = f'roupa={roupa_alvo_2}, comida={comida_alvo_2b:.2f}'

# Inicializar nova figura
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

# Adicionar primeira curva 
ax.plot(roupas, comidas_1, color="red", marker='.', label=r1);
ax.scatter(roupa_alvo, com1, color="black", marker='^', s=50,);

ax.plot(roupas, comidas_2, color="green", marker='+', label=r2);
ax.scatter(roupa_alvo, com2, color="black", marker='^', s=50,);

ax.plot(roupas, comidas_3, color="blue", label=r3);
ax.scatter(roupa_alvo, com3, color="black", marker='^', s=50,);

# Adicionar ponto em roupa=5.5
ax.scatter(roupa_alvo_2, comida_alvo_2a, marker="s", color="black", s=100, label=r4);
ax.scatter(roupa_alvo_2, comida_alvo_2b, marker="o", color="black", s=100, label=r5);

# Configurar presentação 
ax.grid();
ax.set_xlabel("Roupa");
ax.set_ylabel("Comida");
ax.set_ylim([0, max(comidas_1 + comidas_2 + comidas_3)]);
ax.set_xlim([0, max(roupas)]);
ax.set_title('Deslocamentos da FPP');
ax.set_xticks(roupas)
ax.set_yticks(range(0, 101, 10))
ax.legend();
```

## Curva de Demanda

Modelos lineales

$$ D = (m1 \times q) + b1 $$
$$ P = (m2 \times q) + b2 $$

Intersecção de curvas 

$$ D = P $$
$$ (m1 \times q) + b1 = (m2 \times q) + b2 $$

resulta

$$ q = \frac{b2 -b1}{m1 - m2} $$

```python
m1 = -1.2
b1 = 40

m2 = 1.1
b2 = 10

quantidade = range(5, 30, 1)

# Definir funções
d_demanda = lambda x: m1 * x + b1
d_producao = lambda x: m2 * x + b2

demanda = [d_demanda(q) for q in quantidade]
producao = [d_producao(q) for q in quantidade]

# Ponto de intersecção 
q_alvo = (b2 - b1) / (m1 - m2)
p_alvo = d_producao(q_alvo)
r_alvo = f'Equilíbrio P({q_alvo:.2f})={p_alvo:.2f}'

# Inicializar nova figura
fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(1, 1, 1)

ax.plot(quantidade, demanda, color="red", label='Demanda');
ax.plot(quantidade, producao, color="blue", label='Produção');
ax.scatter(q_alvo, d_producao(q_alvo), color="black", marker='o', label=r_alvo);

# Configurar presentação 
ax.set_xlabel("Quantidade");
ax.set_ylabel("Preço");
ax.set_title('Demanda & Produção');
ax.set_ylim([0, max(producao + demanda) + 1]);
ax.set_xlim([0, max(demanda)]);
ax.axvline(x=q_alvo, linestyle=':', linewidth=2)
ax.axhline(y=p_alvo, linestyle=':', linewidth=2)
ax.axhline(y=20, linestyle='-', linewidth=3)
ax.axvspan(xmin=0, xmax=q_alvo, facecolor='k', alpha=0.1)
ax.legend();
```
